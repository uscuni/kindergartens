```python
part3 = part3.rename(columns={'skolné': 'monthly_fee', 'stravné': 'meals'})
part1 = part1.rename(columns={'validated_fee_month': 'monthly_fee'})
part2 = part2.rename(columns={'validated_fee_monthly': 'monthly_fee'})
```
